```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sentiment=pd.read_csv("fear_greed_index.csv")
trades=pd.read_csv("historical_data.csv")
```

```python
sentiment['date']=pd.to_datetime(sentiment['date'])
sentiment['classification']=sentiment['classification'].str.lower()
```

```python
# trades['time']=pd.to_datetime(trades['time'],unit='ms')
# trades['date']=trades['time'].dt.date
# sentiment['date']=sentiment['date'].dt.date

# trades['closedPnL'] = pd.to_numeric(trades['closedPnL'], errors='coerce')
# trades['leverage'] = pd.to_numeric(trades['leverage'], errors='coerce')
# trades['size'] = pd.to_numeric(trades['size'], errors='coerce')

# trades.dropna(subset=['closedPnL','leverage','size'], inplace=True)
trades.columns = (
    trades.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
print(trades.columns)
```

```python
trades = trades.loc[:, ~trades.columns.duplicated()]
```

```python
trades['timestamp'] = pd.to_numeric(trades['timestamp'], errors='coerce')
trades['timestamp'] = pd.to_datetime(trades['timestamp'], unit='ms')

trades['date'] = trades['timestamp'].dt.date
```

```python
num_cols = ['execution', 'size_token', 'size_usd', 'closed_pnl', 'fee', 'start_posit']

for c in num_cols:
    if c in trades.columns:
        trades[c] = pd.to_numeric(trades[c], errors='coerce')
```

```python
trades.isna().sum()
```

```python
trades.dropna(subset=['execution_price','size_tokens','size_usd','closed_pnl'], inplace=True)
```

```python
trades['side'] = trades['side'].str.upper().str.strip()
trades['direction'] = trades['direction'].str.upper().str.strip()
```

```python
trades['is_profit'] = trades['closed_pnl'] > 0
trades['trade_volume'] = trades['size_usd']
```

```python
trades = trades[
    (trades['size_usd'] > 0) &
    (trades['execution_price'] > 0)
]
```

```python
print(trades.columns)
print(sentiment.columns)
```

```python
sentiment.columns = (
    sentiment.columns
    .str.strip()
    .str.lower()
)

sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date
```

```python
sentiment[['date','classification']].head()
```

```python
merged = pd.merge(
    trades,
    sentiment[['date', 'classification']],
    on='date',
    how='inner'
)
```

```python
import os
os.makedirs("datanalyst", exist_ok=True)
```

```python
merged.to_csv("merged_data.csv", index=False)
```

```python
print(merged.columns)
```

```python
merged['is_profit'] = merged['closed_pnl'] > 0
merged['trade_volume'] = merged['size_usd']

merged['risk_level'] = pd.cut(
    merged['size_usd'],
    bins=[0,500,2000,10000,50000,1e9],
    labels=['low','medium','high','very_high','extreme']
)
merged['trade_volume']=merged['execution_price']*merged['size_usd']
```

```python
import os
os.makedirs("outputs", exist_ok=True)
```

```python
#profitability vs sentiment

pnl_sentiment=merged.groupby('classification')['closed_pnl'].mean()

pnl_sentiment.plot(kind='bar')
plt.title('Average PnL by market sentiment')
plt.savefig('outputs/pnl_by_sentiment.png')
plt.show()
```

```python
sns.boxplot(data=merged, x='classification', y='size_usd')
plt.title("size Distribution by Sentiment")
plt.savefig("outputs/leverage_by_sentiment.png")
plt.show()
```

```python
daily_volume=merged.groupby('date')['trade_volume'].sum()

daily_volume.plot()
plt.title("Daily trading volume")
plt.savefig("outputs/volume_trade.png")
plt.show()
```

```python
corr=merged[['closed_pnl','size_usd','trade_volume']].corr()

sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.savefig("outputs/correlation_heatmap.png")
plt.show()
```

```python
daily_metrics=merged.groupby(['date','classification']).agg({
    'closed_pnl':'mean',
    'size_usd':'mean',
    'size_tokens':'mean',
    'trade_volume':'sum'
}).reset_index()

daily_metrics.to_csv("datanalyst/daily_metrics.csv", index=False)
    
```

```python
side_stats = merged.groupby(['classification','side'])['closed_pnl'].mean().unstack()

side_stats.plot(kind='bar')
plt.title("PnL by Trade Side and Sentiment")
plt.savefig("outputs/pnl_side_sentiment.png")
plt.show()
```

```python
side_stats = merged.groupby(['classification','side'])['closed_pnl'].mean().unstack()

side_stats.plot(kind='bar')
plt.title("PnL by Trade Side and Sentiment")
plt.savefig("outputs/pnl_side_sentiment.png")
plt.show()
```

```python
summary = merged.groupby('classification').agg({
    'closed_pnl':['mean','std'],
    'size_usd':'mean',
    'trade_volume':'sum'
})

summary.to_csv("datanalyst/summary_by_sentiment.csv")
```
